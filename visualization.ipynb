### ৩.১.৫ বর্ননামূলক পরিসংখ্যানের মাধ্যমে ডেটার সম্পর্কে জানা

বর্ননামূলক পরিসংখ্যানের(Descriptive statistics) মাধ্যমে ডেটার ভেতরকার খবর দ্রুত এবং ভালোভাবে জানা যায়। pandas এর describe() ফাংশনের মাধ্যমে প্রতিটি ফিচারের ৮ টি statistical properties দেখা যায়। 

১) count ২) Mean ৩) Standard deviation ৪) Minimum value ৫) 25th percentile ৬) 50th percentile ৭) 75th percentile ৮) Maximum value

```python
# Descriptive statistics
dataset.describe()
```

count দ্বারা আমরা জানতে পারি কতগুলো অবজারভেশন রয়েছে ঐ ফিচারের। এর মাধ্যমে কোন ফিচারের ডেটা মিসিং আছে কি না জানা যায়। ডেটা মিসিং থাকলে তা ডেটা প্রিপ্রসেসিং টেকনিকের মাধ্যমে তা হ্যান্ডেল করতে হবে। 
আমরা দেখতে পাচ্ছি এখানে কোন ফিচারের ডেটা মিসিং নেই। 

Mean হচ্ছে ঐ ফিচারের গড় ভ্যালু। যা ঐ ফিচারের ডেটার প্রতিনিধিত্ব করে।

std হচ্ছে standard deviation যার মাধ্যমে আমরা ডেটার ভ্যারিয়েশন জানতে পারি। std বেশি হলে বুঝতে হবে ঐ ফিচারের ভ্যালুগুলোর বিস্তার অর্থাৎ একটি ভ্যালুর থেকে আরেকটি ভ্যালুর দূরত্ব বেশি।  এক্ষেত্রে ডেটা প্রসেসিং টেকনিক ব্যবহার করে ব্যপারটি হ্যান্ডেল করতে হবে। std এর মানে -3<std<3 এর মধ্যে হলে ভালো। 

min এবং max এর ফিচারের ভ্যালুর সবচেয়ে ছোট এবং সবেচেয়ে বেশি মান নির্দেশ করে। এর মাধ্যমে ডেটার রেঞ্জ সম্পর্কে ধারনা পাওয়া যায়। 

25%, 50%, 75% এর মাধ্যমে জানা যায় ভ্যালুগুলো এর নিচে অবস্থান করছে। যেমন  আমাদের ডেটাসেটের Pregnancies ফিচারের 25% ভ্যালু 1.000000 এর নিচে অবস্থান করছে। 

তাহলে আমরা Descriptive statistics এর মাধ্যমে ডেটা সম্পর্কে গভীর ধারনা পাই এবং কোন কোন ফিচারকে প্রিপ্রসেসিং করতে হবে  সেই সিদ্বান্ত নেওয়া যায়। 

## ৩.১.৬ ক্লাস ডিস্ট্রিবিউশন সম্পর্কে জানা 


ক্লাসিফিকেশন সমস্যায় আমাদের লেভেল ডেটা হিসেবে হ্যাঁ অথবা না থাকে অর্থাৎ 0 অথবা 1 থাকে (মাল্টিপল ক্লাসিফিকেশন প্রবলেম এ ক্ষেত্রে 0, 1, 2 থাকে)। ক্লাস ডিস্ট্রিবিউশন হচ্ছে আমাদের ডেটা সেটে কয়টি 0 এবং 1 আছে সে তা জানা । এটি জানা জরুরী কেননা যদি ক্লাস ডিস্ট্রিবিউশনে যদি অনেক বেশি পার্থক্য থাকে তাহলে তা মেশিনকে শেখানোর সময় খারাপ প্রভাব ফেলতে পারে। সেক্ষেত্রে ডেটা প্রিপ্রসেসিং এর সময়  স্পেশাল টেকনিকের মাধ্যমে ডেটাকে হ্যান্ডেল করতে হবে। 

```python
class_counts = dataset.groupby('Outcome').size()
print(class_counts)
```

আমরা এখানে দেখতে পাচ্ছি 0 ক্লাসে অবজারভেশন 1 ক্লাসের অবজারভেশনের প্রায় দ্বিগুণ 

## ৩.১.৬ ফিচারগুলোর মধ্যকার সম্পর্ক (Correlation)

কোরিলেশন হচ্ছে একটি ভ্যারিয়েবলের পরিবর্তনের ফলে আরেকটি ভ্যালু কিরূপ পরিবতর্ন হয় । এই সম্পর্কে অধ্যায় দুইইয়ে বিস্তারিত আলোচনা করা হয়েছে। 
ফিচারগুলোর মধ্যে কোরিলেশন জানা জরুরী কেননা ভ্যারিয়েবলগুলোর মধ্যে যদি highly correlation থাকে তাহলে কিছু  কিছু মেশিন লার্নিং এলগোরিদম (linear & Logistic regression) আছে  যেগুলো ডেটা থেকে ভালো ভালো শিখতে পারে না। 

```python
dataset.corr()
```

 আমরা এখানে দেখতে পাচ্ছি ডেটাগুলো নিজেদের সাথে ছাড়া অন্য কোন এট্রিবিউটের সাথে Highly correlation নাই। যদি অন্য কোন ফিচারের সাথে  -0.8 থেকে -1.0 থাকত তাহলে তাকে Negative highly correlation বলা হত । অথবা 8.0 থেকে 1.0 হত তাহলে Positive highly correlation বলা হত। 

## ৩.১.৭  এট্রিবিউটগুলো skew সম্পর্কে জানা 

নরমাল বা গসিয়ান ডিস্ট্রিবিউশনের কিছুটা পরিবর্তনই হচ্ছে skew। অনেক মেশিন লার্নিং এলগরিদম আছে যেগুলো ডেটার নরমাল ডিস্ট্রিবিউশন থেকে ডেটা সম্পর্কে ভালোভাবে শিখতে পারে। তাই Skew সম্পর্কে জেনে ডেটা প্রিপ্রসেসিং টেকনিক ব্যবহার করে তাদেরকে নরমাল ডিস্ট্রিবিউশনে পরিবর্তন করতে হবে এতে করে মেশিন লার্নিং মডেলের পারফর্মেন্স ভালো হবে।
