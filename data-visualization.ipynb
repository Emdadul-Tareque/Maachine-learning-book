# ৪.১ ইউনিভ্যারিয়েট প্লট (Univariate Plots)

Univariate Plots  হচ্ছে প্রতিটা ফিচারকে আলাদা আলাদাভাবে প্লট করা। অর্থাৎ একটি লেখচিত্রে শুধুমাত্র একটি ভ্যারিয়েবলের ডেটাগুলো প্লট হবে। Univariate Plots এর মাধ্যমে প্রতিটা ফিচারের ডেটা ডিস্ট্রিবিউশন সম্পর্কে ধারনা পাওয়া যায়। ডেটার বিস্তার এবং কোন কোন ডেটাগুলো  outlier অর্থাৎ দলছুট ভ্যালু তা জানা যায়। দলছুট ভ্যালু ডেটার পুরো সামারাইজেশন চেঞ্জ করে দেয়। তাই দলছুট ভ্যালুগুলোকে ডেটা থেকে ফেলে দিতে হবে। Outlier অর্থাৎ দলছুট ভ্যালু হচ্ছে ঐ ভ্যালুগুলো যেগুলা একটা নির্দিস্ট একত্রিত ডেটার থেকে অনেক দূরে অবস্থান করে যেমনঃ মনে করি ক্লাস ৫ এ পড়া একটা ছাত্রের বয়স ১০ থেকে ১১ বছরের মধ্যে হবে এখন যদি সেই ক্লাসে ৩০ বছরের এক আদুভাই পড়ে তাহলে এই আদুভাই হচ্ছে দলছুট ভ্যালু বা outlier। 

আমার এখানের তিনটি Univariate Plots টেকনিক নিয়ে আলোচনা করব --

ক) হিস্টোগ্রাম প্লট (Histogram plots)

খ) ডেনসিটি প্লট (Density plots)

গ) বক্স এবং হুইস্কার প্লট (Box & Whisker Plots)


## ৪.১.১ হিস্টোগ্রাম প্লট (Histogram plots)

প্রতিটা ফিচারের ডিস্ট্রিবিউশন আলাদা আলাদাভাবে দেখার জন্য  হিস্টোগ্রাম প্লট (Histogram plots) একটি ভালো টেকনিক। হিস্ট্রোগ্রাম গ্রুপ ডেটাকে একটা Bin এ নিয়ে আসে এবং প্রতিটা bin এর কয়টা অবজারভেশন আছে তা কাউট করে প্লট করে Bar আকারে। এই bar এর আকার থেকে বুঝা যায় এই ফিচারের ডেটা ডিস্ট্রিবিউশন কোন ধরনের (নারমাল ডিস্ট্রিবিউশন, skew, এক্সপোনেনশিয়াল (exponential)। 

Pandas DataFrame এর মাধ্যমে ডেটা ভিজ্যুয়ালাইজেশন করা যায়। hist() ফাংশন ব্যবহার করে হিস্টোগ্রাম প্লট (Histogram plots) করা যায় । 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('diabetes.csv')
sns.set(style='darkgrid')
dataset.hist(bins=50, figsize = (15, 10))
plt.show()
```

আমরা histogram থেকে দেখতে পাচ্ছি age, DiabetesPedigreeFunction, Insulin, Pregancies এর ডেটা ডিস্ট্রিবিউশন হচ্ছে exponential distribution। এবং BMI, Blood pressure, Glucose এর ডেটা ডিস্ট্রিবিউশন নরমাল / গসিয়ান ডিস্ট্রিবিউশন বা এর কাছাকাছি ডিস্ট্রিবিউশন। 

## ৪.১.২ ডেনসিটি প্লট (Density plots)

ডেটা ডিস্ট্রিবিউশন সম্পর্কে জানার জন্য আরেকটি ভালো টেকনিক হচ্ছে ডেনসিটি প্লট (Density plots)। 
পাইথনে কিভাবে ডেনসিটি প্লট (Density plots) আঁকতে হয় দেখা যাক

```python
dataset.plot(kind='kde', subplots=True, layout=(3,3), sharex=False, figsize=(15, 10))
plt.show()
```

histogram plot  থেকে  density plot এ ডেটা ডিস্ট্রিবিউশন আরো ভালোভাবে দেখা যাচ্ছে। 
আমরা Density plot থেকে দেখতে পাচ্ছি age, DiabetesPedigreeFunction, Insulin, Pregancies এর ডেটা ডিস্ট্রিবিউশন হচ্ছে exponential distribution। এবং BMI, Blood pressure, Glucose এর ডেটা ডিস্ট্রিবিউশন নরমাল / গসিয়ান ডিস্ট্রিবিউশন বা এর কাছাকাছি ডিস্ট্রিবিউশন। 

# ৪.১.৩  বক্স এবং হুইস্কার প্লট (Box and Whisker Plots)

বক্স এবং হুইস্কার প্লটের মাধ্যমে ডেটার বিস্তার সম্পর্কে ধারনা পাওয়া যায়। বক্সপ্লটের মাধ্যমে ডেটার skew সম্পর্কেও পরিষ্কার ধারনা পাওয়া যায়। ডেটার outlier অর্থাৎ দলছুট ভ্যালু সম্পর্কে স্পষ্ট ধারনা পাওয়া যায়। 
আমরা এখানে cufflinks এর মাধ্যমে বক্সপ্লট আঁকব।

```python
import cufflinks as cf
cf.go_offline()

fig = dataset.iplot(asFigure=True, kind="box")
fig.show()
```

আমরা বক্স অ হুইস্কার প্লট থেকে দেখতে পাচ্ছি ইনসুলিনের ডেটাগুলো বিস্তার অনেক বেশি বিশেষ করে q3 থেকে ম্যাক্সিমাম পর্যন্ত। এবং অন্যান্য ফিচারের বক্স প্লট থেকে ডেটার skew সম্পর্কে একটা ধারনা পাওয়া যাচ্ছে। অধ্যায় ২ এ বক্স এবং হুইস্কার প্লট নিয়ে বিস্তারিত আলোচনা করা হয়েছে।

## ৪.২.১ কোরিলেশন ম্যাট্রিক্স প্লট (Correlation metrix plots)

কোরিলেশন হচ্ছে একটি ভ্যারিয়েবলের পরিবর্তনের ফলে আরেকটি ভ্যালু কিরূপ পরিবতর্ন হয় । এই সম্পর্কে অধ্যায় দুইইয়ে বিস্তারিত আলোচনা করা হয়েছে। ফিচারগুলোর মধ্যে কোরিলেশন জানা জরুরী কেননা ভ্যারিয়েবলগুলোর মধ্যে যদি highly correlation থাকে তাহলে কিছু কিছু মেশিন লার্নিং এলগোরিদম (linear & Logistic regression) আছে যেগুলো ডেটা থেকে ভালো ভালো শিখতে পারে না।
আমরা seaborn ব্যবহার করে Correlation metrix plots আঁকব। 

```python
corr = dataset.corr()
fig = plt.figure(figsize=(15,6))
sns.heatmap(corr, linecolor='black', linewidths=.1, annot=True)
```

আমরা Correlation metrix plots থেকে দেখতে পাচ্ছি এট্রিবিউটগুলোর মধ্যে(নিজের সাথে ছাড়া )  highly correlation নেই। highly correlation থাকলে খয়েরি কালার হতো। 

# ৪.২.২ স্ক্যাটার প্লট ম্যাট্রিক্স (Scatter Plot Matrix)

স্ক্যাটার প্লটের মাধ্যমে দুটি নিউমেরিক্যাল ভ্যারিয়েবলের মধ্যে কোরিলেশন এবং দলছুট ভ্যালুগুলোকে দেখা যায়। ডেটা সেটের প্রতিটা ভ্যারিয়েবলের সাথে অপর আরেকটি ভ্যারিয়েবলের স্ক্যাটার প্লট করে তাদের মধ্যে সম্পর্ক দেখা যায়। এই সবগুলো প্লট যখন একসাথে দেখানো হয় তখন তাকে স্ক্যাটার প্লট ম্যাট্রিক্স (Scatter Plot Matrix) বলে। 
আমরা seaborn ব্যবহার করে স্ক্যাটার প্লট ম্যাট্রিক্স (Scatter Plot Matrix) তৈরি করব। 

```python
sns.pairplot(data=dataset, hue='Outcome')
```

আমরা এখানে দেখতে পাচ্ছি ভ্যারিয়েবলগুলোর মধ্যে কোন স্ট্রং কোরিলেশন নেই।
