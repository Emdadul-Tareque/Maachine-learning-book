# অধ্যায় ২ঃ একটু খানি পরিসংখ্যান 

ডেটার ভেতকার খবর জানতে অর্থাৎ ডেটা সম্পর্কে ভালো ধারনা নিতে দরকার পরিসংখ্যান (statistics)। এই অধ্যায়ে আমরা  পরিসংখ্যানের বেসিক কিছু জিনিস নিয়ে আলোচনা করব। 

## ২.১ কেন্দ্রীয় প্রবণতা ও পরিমাপ 

তথ্যের গাণিতিক বৈশিষ্ট্য জানার জন্য কেন্দ্রীয় প্রবণতা ও কেন্দ্রীয় প্রবণতার পরিমাপ জানা প্রয়োজন। তথ্যসমূহের যথাযথ বিশ্লেষনের জন্য সংখ্যাগত বিশ্লেষণ জানা প্রয়োজন। সংখ্যাগত বিশ্লেষণ জানতে পারলে তথ্যসমূহকে একটি প্রতিনিধিত্বমূলক সংখ্যা দ্বারা প্রকাশ করা যায়। এই জন্য কেন্দ্রীয় প্রবনতার পরিমাপ সম্পর্কে জানা দরকারি। 

## ২.১.১ কেন্দ্রীয় প্রবণতা 

কোন একটি নিবেশন  বা তথ্য সারিতে অনেকগুলো মান থাকে। আর একটি মান কেন্দ্রে থাকে। কেন্দ্রীয় মানের আশেপাশে অন্যান্য ডেটা একত্রিত বা ঘনীভূত থাকতে চায় । ডেটার এই মাঝামাঝি থাকার প্রবনতাকে কেন্দ্রীয় প্রবনতা বলে 

## ২.১.২ কেন্দ্রীয় প্রবণতার পরিমাপ 

যে সকল গাণিতিক পরিমাপের সাহায্যে কেন্দ্রীয় মান নির্নয় করা যায় তাকে কেন্দ্রীয় প্রবণতার পরিমাপ বলা হয়। কেন্দ্রীয় প্রবণতার পরিমাপ পাঁচ প্রকার -- 

১) গাণিতিক গড় (Arithmetic mean)

২) জ্যামিতিক গড় (Geometric mean)

৩) তরঙ্গ গড় (Harmonic mean)

৪) মধ্যক (median)

৫) প্রচুরক (Mode)

মেশিন লার্নিয়ে সাধারণত গাণিতিক গড়, মধ্যক ও প্রচুরক কাজে লাগে। 

১) গাণিতিক গড় (Arithmetic mean): ডেটার সবগুলো মানের যোগফল বা সমষ্টিকে মোট ডেটার সংখ্যা দিয়ে ভাগ করলে গাণিতিক গড় পাওয়া যায়। গাণিতিক গড় প্রান্তীয় মান দ্বারা প্রভাবিত হয় তাই গাণিতিক গড় দিয়ে ডেটা সম্পর্কে জানার ক্ষেত্রে সতর্ক থাকতে হয়। 

২) মধ্যক (Median): তথ্যের সারণীর মাঝের সংখ্যাটি হচ্ছে মধ্যক। যেমন; ৫, ৯, ১৩, ১৫, ১৭। এই সারণীর মধ্যক হচ্ছে ১৩। 

৩) প্রচুরক (Mode): তথ্যের সারণীর মধ্যে যে সংখ্যাটা সবচেয়ে বেশি থাকে তাকে প্রচুরক বলে। যেমনঃ ৫, ৮, ৭, ৭, ৯, ১১, ১৩। 
এখানে ৭ সংখ্যাটি হচ্ছে প্রচুরক। 

# ২.২ ডেটার বিস্তার (Dispersion)

কেন্দ্রীয় মান থেকে Distribution এর অন্যান্য মানগুলোর ব্যবধানকে বিস্তার বলে। এর মাধ্যমে ডিস্ট্রিবিউশনের কেন্দ্রীয় মান থেকে অন্যান্য মানগুলো কতদূরে অবস্থান করছে শয়ে সম্পর্কে ধারনা পাওয়া যায়। 

অর্থাৎ Distribution এর গড়/মধ্যকের মান হতে উহার অন্যান্য সংখ্যা কত ব্যর বা ছোট তার পরিমাপকে বিস্তার বলে। কোন Distribution ডেটাগুলো যত বিক্ষিপ্ত থাকে উহাদের বিস্তার তত বাড়তে থাকে। 

## ২.২.১ বিস্তার পরিমাপের প্রকারভেদ 

বিস্তারকে চারভাবে পরিমাপ  করা যায় -- 

১) পরিসর (range)

২) চতুর্থক ব্যবধান (Fourth Quartile)

৩) গড় ব্যবধান 

৪) পরিমিতি ব্যবধান (Standard deviation)

মেশিন লার্নিং এ চতুর্থক ব্যবধান ও পিরিমির ব্যবধান সম্পর্কে জানা প্রয়োজন তাই আমরা এখানে এই দুইটা নিয়ে আলোচনা করব। 

###  চতুর্থক ব্যবধান (Fourth Quartile)

কোন তথ্য সারির মানগুলোকে মানের উর্ধক্রম বা নিম্মক্রমে সাজানোর পর যে মানগুলো ঐ তথ্য সারিকে চারভাগে ভাগ করে তাদেরকে চতুর্থক বলে। 

বক্সপ্লট জানার জন্য চতুর্থক ব্যবধান জানা প্রয়োজন । আসুন আমরা বক্সপ্লট সম্পর্কে জানি ।
